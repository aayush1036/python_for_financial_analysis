```
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,11)
y = x**2
```

```
x
```

```
y
```

```
#functional method
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title');
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```
#Object Oriented Method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])  #left, bottom, width, height as a percentage of total blank canvas
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) 
#20% from left, 50% from bottom, 40% width in proportion to empty canvas, 30% height in proportion to empty canvas
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes1.set_title('Larger plot');
```

```
fig,axes = plt.subplots(nrows=1, ncols=2) #creating a plot grid fo 1 row and 2 columns and filling it with plots
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
plt.tight_layout()
```

Figsize and DPI

```
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(5,3)) #creating a plot grid of 2 rows and 1 columns and filling it with plots and setting the 
#figure size to 5 inches wide and 3 inches tall
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout() #ensures no overlapping
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X squared')
ax.plot(x,x**3, label='X cubed')
ax.legend(loc=0); #the loc parameter helps you to set the location of the legend. Pass a tuple of co ordinates just like in 
#fig.add_axes() and it places the legend there
```
