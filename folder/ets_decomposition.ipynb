```
import pandas as pd
import matplotlib.pyplot as plt
```

```
airline = pd.read_csv('airline_passengers.csv', index_col='Month')
airline.dropna(inplace=True)
airline.index = pd.to_datetime(airline.index)
```

```
airline.head()
```

```
airline.plot()
```

```
from statsmodels.tsa.seasonal import seasonal_decompose
```

```
result = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')
result.plot()
#we are using multiplicative model since the growth looks like an exponential function and not a linear function
#if the trend looks like it is linear then we can use the additive model
```

```
# You may accidentally see two of the same plots here, not to worry,
# just a small bug with statsmodels function.
```

```
#to access individual components we can use the attributes 
result.seasonal.plot()
```

```
result.trend.plot()
```
